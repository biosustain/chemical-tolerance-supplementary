```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.manifold import MDS, TSNE
import seaborn as sbn
```

```python
%matplotlib inline
```

```python
full_df = pd.read_csv("/Users/krisj/Downloads/distances.tsv", sep="\t")
df = full_df.groupby("molecule").mean().transpose()

comp_name_order = [
    "1,2-propanediol", "2,3-butanediol", "hexamethylenediamine", "putrescine", "glutaric acid",
    "adipic acid", "hexanoic acid", "octanoic acid", "coumaric acid", "isobutyric acid", "butanol"
]
df = df[comp_name_order]
```

```python
plt.plot(df["2,3-butanediol"], df["1,2-propanediol"], ".")
```

```python
method = list(set(full_df["fingerprint_format"]))[2]
print(method)
plot_df = full_df[full_df["fingerprint_format"] == method].set_index("molecule")[df.index].transpose()[comp_name_order]
sbn.heatmap(plot_df.corr())
```

```python
pca = MDS(2)
coords = pca.fit_transform(df)
```

```python
plt.plot(coords[:, 0], coords[:, 1], ".")
```
